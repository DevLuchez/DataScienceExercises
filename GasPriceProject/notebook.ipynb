```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Armazenamento dos dados no dataframe
dfGas = pd.read_csv('datasets\gas_prices.csv')
dfGas.head()
```

```python
# Limpeza dos dados
dfGas.dropna()
dfGas
```

```python
# Definição do título do gráfico
plt.title('Valores do Combustível em US$')

# Criação de gráficos de linha, definição da label , da cor e do formato (pontilhado, tracejado) de cada linha 
plt.plot(dfGas['Year'], dfGas['Australia'], 'r', label='Australia')
plt.plot(dfGas['Year'], dfGas['Italy'], 'y-', label='Italy')
plt.plot(dfGas['Year'], dfGas['USA'], 'b.', label='USA')
plt.plot(dfGas['Year'], dfGas['Japan'], 'g.-', label='Japan')

# Ajustando períodos do eixo X, do início ao fim com um intervalo de 2 em 2 anos
plt.xticks(dfGas['Year'][::2])

# Definindo título dos eixos X e Y
plt.xlabel('Years')
plt.ylabel('Gas Price (US$)')

# Plotar a legenda e o gráfico
plt.legend()
plt.savefig('Gas Price Graph.jpeg') # exportando gráfico em PNG
plt.show() # ao mostrar, zera o gráfico e a imagem acima fica em branco
```
