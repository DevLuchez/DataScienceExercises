```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Armazenamento dos dados no dataframe
dfGas = pd.read_csv('datasets\gas_prices.csv')
dfGas.head()
```

```python
# Limpeza dos dados
dfGas.dropna()
dfGas
```

```python
# Definio do ttulo do grfico
plt.title('Valores do Combustvel em US$')

# Criao de grficos de linha, definio da label , da cor e do formato (pontilhado, tracejado) de cada linha 
plt.plot(dfGas['Year'], dfGas['Australia'], 'r', label='Australia')
plt.plot(dfGas['Year'], dfGas['Italy'], 'y-', label='Italy')
plt.plot(dfGas['Year'], dfGas['USA'], 'b.', label='USA')
plt.plot(dfGas['Year'], dfGas['Japan'], 'g.-', label='Japan')

# Ajustando perodos do eixo X, do incio ao fim com um intervalo de 2 em 2 anos
plt.xticks(dfGas['Year'][::2])

# Definindo ttulo dos eixos X e Y
plt.xlabel('Years')
plt.ylabel('Gas Price (US$)')

# Plotar a legenda e o grfico
plt.legend()
plt.savefig('Gas Price Graph.jpeg') # exportando grfico em PNG
plt.show() # ao mostrar, zera o grfico e a imagem acima fica em branco
```
