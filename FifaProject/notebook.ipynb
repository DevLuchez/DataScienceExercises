```python
import pandas as pd
```

```python
# Ler e exibir os dados do arquivo CSV fifa
dados = pd.read_csv('datasets/fifa.csv')

dados.head()
```

```python
# Exibir as colunas presentes no df dados
dados.columns
```

```python
# Consultar o nome e a nacionalidade dos 3 primeiros registros do df dados
dados.head(3)[['Name', 'Nationality']]

# Consultar o nome e a nacionalidade dos 3 últimos registrs do df dados
dados.tail(3)[['Name', 'Nationality']]
```

```python
# Localizar os registros do index/row 1 ao antecedente do 5
dados.iloc[1:5]
```

```python
# Localizar o registro da linha 2, coluna 2 
dados.iloc[2,2]
```

```python
# Localizar, entre os primeiros 50 registros, aqueles que tiverem a coluna Nationality igual a Brazil
dados.head(50).loc[dados['Nationality'] == 'Brazil']
```

```python
# Consultar os registros ordenando o Age e o Name decrescente
dados.sort_values(['Age', 'Name'], ascending=False)
```

```python
# Consultando o Name e o Wight contidos no df dados, ordenando decrescente pelo peso
dados[['Name', 'Weight']].sort_values('Weight', ascending=False)
```

```python
# Adicionando nova coluna no df dados
dados['best_player_classification'] = dados['Acceleration'] + dados['Agility'] + dados['Reactions']
dados
```

```python
# Mudando o nome da coluna de classificação total
dados.rename(columns={'best_player_classification': 'BestPlayerClassification'}, inplace=True)
dados
```

```python
# Criando novo df para consultar os jogadores com a melhor classificação
dfBestPlayers = dados[['Name', 'BestPlayerClassification']]
dfBestPlayers.sort_values(['BestPlayerClassification'], ascending=False).head()
```

```python
# Exportando df para um arquivo csv, tirando os IDs do df dados
dfBestPlayers.to_csv('CSV_BestPlayers.csv', index=False)
```
