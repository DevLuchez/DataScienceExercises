```python
import pandas as pd
```

```python
dfMelbourne = pd.read_csv('datasets/MelbourneHouseData.csv')
dfMelbourne
```

```python
# Limpando preÃ§os nulos
dfMelbourne.dropna(subset='Price')
dfMelbourne.drop_duplicates()
dfMelbourne.sort_values(['Address'], ascending=True).head(3)
```

```python
# EstatÃ­sticas Descritivas bÃ¡sicas
dfMelbourne[['Price', 'Rooms', 'Distance']].describe().loc[['mean', 'count', 'std']]
```

```python
# 10 bairros com o maior nÃºmero de imÃ³veis Ã  venda
dfMelbourne['Suburb'].value_counts().head(10)
```

```python
dfFilter_1 = dfMelbourne.loc[(dfMelbourne['Rooms'] == 3) & (dfMelbourne['Type'] == 'h') & (dfMelbourne['Price'] <= 300000)]
dfFilter_1[['Rooms', 'Type', 'Price']].sort_values(['Price'], ascending=False)
```

```python
import re

dfFilter_2 = dfMelbourne.loc[(dfMelbourne['Address'].str.contains('Turner St|Turner Rd', flags=re.I)) & (dfMelbourne['Method'] == 'S')]
dfFilter_2[['Address', 'Method']]
```

```python
dfFilter_3 = dfMelbourne.loc[(dfMelbourne['Address'].str.contains('^59', flags=re.I)) & (dfMelbourne['Price'] <= 500000)]
dfFilter_3[['Address', 'Price', 'Method']].sort_values(['Address'], ascending=True)
```
